### 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(database):
    from env import user, password, host
    return f'mysql+pymysql://{user}:{password}@{host}/{database}'
```

```python
url = get_db_url('chipotle')
query = "SELECT * FROM orders"
orders = pd.read_sql(query, url)
orders.head()
```

```python
# make a variabel for the top three items
top_four_items = orders.groupby('item_name').quantity.agg('sum').sort_values(ascending=False).head(4)
top_four_items
```

```python
top_four_items.index
```

```python
orders.info()
```

```python
# Convert the item_price column into a float
orders['item_price'] = orders.item_price.str.replace('$', '').astype('float')
```

```python
orders.head()
```

```python
# revenue produced by 4 highest selling items.

revenue = orders[orders.item_name.isin(top_four_items.index)].groupby('item_name').item_price.sum()
revenue
```

```python
revenue.plot(kind='bar')
```

```python
# another way
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
plt.figure(figsize = (8, 5))
sns.barplot(y = 'item_name', x = 'item_price', data = revenue)
```
