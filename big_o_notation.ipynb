```python
import matplotlib.pyplot as plt
import math
import numpy
```

```python
plt.figure(figsize = (12,10))

y = [1 for n in x]
plt.plot(x, y, label = '$O(1)$')


x = (range(1, 8))
y = [math.log(n) for n in x]
plt.plot(x, y, label = '$Olog(n)$')


y = [n for n in x]
plt.plot(x, y, label = '$O(n)$')


y = [n * math.log(n) for n in x]
plt.plot(x, y, label = '$O(n)logn$')


y = [n**2 for n in x]
plt.plot(x, y, label = '$O(n^2)$')


y = [2**n for n in x]
plt.plot(x, y, label = '$O(2^n)$')



y = [math.factorial(n) for n in x]
plt.plot(x, y, label = '$O(n!)$')
plt.ylim(-10, 100)


y = [n**n for n in x]
plt.plot(x, y, label = '$O(n^n)$')


plt.xlabel('Elements')
plt.ylabel('Operations')

plt.legend()
plt.show()
```
