```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# 1. Use matplotlib to plot the following equation:

# y = x^2 - x + 2

# You'll need to write the code that generates the x and y points.

# Add an anotation for the point 0, 0, the origin.
```

```python
x = (range(-30, 31))
y = [n**2 - n + 2 for n in x]
plt.plot(x, y)
plt.ylim(-5, 100)
plt.text(0,0, '(0,0)', fontsize=15, color='red', ha='center')

plt.show()
```

```python
# 2. Create and label 4 separate charts for the following equations 
# (choose a range for x that makes sense):
```

```python
x = (range(1, 150))
y = [math.sqrt(n) for n in x]
plt.plot(x, y)
plt.ylim(-5, 50)
plt.text(40, 8, '$y= sqrt(x)$', fontsize=10, color='red')

plt.show()
```

```python
# for y = x^3
```

```python
x = (range(-500, 500))
y = [n**3 for n in x]
plt.plot(x, y)
plt.text(.25, 50, '$y= x^3$', fontsize=10, color='red')

plt.show()
```

```python
# for y = 2^x
```

```python
x3 = (range(-100, 10))
y3 = [2**n for n in x3]
plt.plot(x3, y3)
plt.text(-8, 25, '$y= 2^x$', fontsize=10, color='red')

plt.show()
```

```python
# for y = 1/(x+1)
```

```python
x1 = (range(-50, -2))
y1 = [(1/(n+1)) for n in x1]

x2 = (range(0, 50))
y2 = [(1/(n+1)) for n in x2]

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.text(-6, 0, '$y = 1/(x+1)$', fontsize=10, color='red')

plt.show()
```

```python
plt.figure(figsize = (12,10))
plt.subplot(2, 2, 1)
x = (range(1, 150))
y = [math.sqrt(n) for n in x]
plt.plot(x, y)
plt.ylim(-5, 50)
plt.text(40, 8, '$y= sqrt(x)$', fontsize=10, color='red')



plt.subplot(2, 2, 2)
x = (range(-500, 500))
y = [n**3 for n in x]
plt.plot(x, y)
plt.text(.25, 50, '$y= x^3$', fontsize=10, color='red')




plt.subplot(2, 2, 3)
x3 = (range(-100, 10))
y3 = [2**n for n in x3]
plt.plot(x3, y3)
plt.text(-8, 25, '$y= 2^x$', fontsize=10, color='red')




plt.subplot(2, 2, 4)
x1 = (range(-50, -2))
y1 = [(1/(n+1)) for n in x1]

x2 = (range(0, 50))
y2 = [(1/(n+1)) for n in x2]

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.text(-6, 0, '$y = 1/(x+1)$', fontsize=10, color='red')
plt.show()
```
